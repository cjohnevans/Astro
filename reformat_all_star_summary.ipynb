```python
import os
import matplotlib.pyplot as plt
```

```python
in_file = '/home/john/astro/variable_star_data/baa/all_star_summary.txt'
out_file = '/home/john/astro/variable_star_data/baa/all_star_summary.csv'
```

```python
f = open(in_file, 'r')
f_out = open(out_file, 'w')
```

```python
# header
hdr = f.readline().replace('\t',',')
print(hdr)
f_out.write(hdr)
```

```python
line3 = 'a'
num_fields = []
while line3 != '':
    line1 = f.readline().replace('\n',' ').replace(',','')
    line2 = f.readline().replace('\n',' ').replace(',','')
    line3 = f.readline().replace('\n',' ').replace(',','')
    full_line =  (line1 +  line2 + line3).split('\t')
    full_line_out = ','.join(full_line) + '\n'
    print(full_line_out)
    f_out.write(full_line_out)
    num_fields.append(len(full_line))
```

```python
f.close()
```

```python
plt.plot(num_fields)
```

```python
# check the output file
f_check = open(out_file,'r')
```

```python
check_line = '1'
n_el = []
while check_line != '':
    check_line = f_check.readline()
    n_el.append(len(check_line.split(',')))
```

```python
f_check.close()
```

```python
plt.plot(n_el)
```
